```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,100)

def sine(x):
    return np.sin(2 * np.pi * x)      #dont forget the (2 *np.pi * x) when working with sin & cosine
def cosine(x):
    return np.cos(2 * np.pi * x)

```

```python
# this creates a multipanel plot with 1 row and 2 columns

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4))

#This sets the x vs sin(x) panel

ax1.plot(x, sine(x))
ax1.set_title('sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid(True)

#This sets the cos(x) panel vs x

ax2.plot(x, cosine(x), color='orange')
ax2.set_title('cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.grid(True)

plt.tight_layout()

plt.savefig('Session8_figure.pdf')
```
