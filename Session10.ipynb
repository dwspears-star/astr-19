```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rv = np.random.default_rng()

random_integers = rv.integers(1, 1000, size=1000)
print(f"Random integers: {random_integers}")
```

```python
plt.hist(random_integers, bins=100)

plt.xlabel('x')
plt.ylabel('y')

plt.savefig("Random_Integers10.pdf")
```
