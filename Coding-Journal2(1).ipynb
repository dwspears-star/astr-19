```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)    #creates an array ranging from 0,1,100

def exp_function(array):      #exponent function that takes that array and returns it in the define function.
    return np.exp(array)
```

```python
y = exp_function(x) #makes y = x

plt.plot(x,y) #this plots the variables x, and y

plt.xlabel("Time[milliseconds]")
plt.ylabel("Awesomeness")

plt.savefig("Coding-Journal-7.pdf")
```
